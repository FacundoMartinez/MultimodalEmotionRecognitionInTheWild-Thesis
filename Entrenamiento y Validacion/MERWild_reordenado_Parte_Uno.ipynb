## Carga y preproceamiento

```python
class ImageDataset(Dataset):
    def __init__(self, csvfile, ruta_BD, root="", transform=None):
        super(ImageDataset, self).__init__()
        df_0 = pd.read_csv(csvfile)
        df = df_0[df_0['Image_Filter'] == 1].copy()  
        df['Filepath'] = ruta_BD + "\\" + df['KEY']
        # Porcentaje deseado (10%)
#         percentage = 0.10
#         # Utiliza el método 'sample' para obtener una muestra aleatoria del 10%
#         df = df.sample(frac=percentage, random_state=42)
        
        existing_files = [os.path.join(root, impath) for impath in df['Filepath'].tolist() if os.path.exists(os.path.join(root, impath))]

        df = df[df['Filepath'].isin(existing_files)]
           
        self.dataframe = df
        self.transform = transform
        self.root = root

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, index):
        if torch.is_tensor(index):
            index = index.tolist()

        sample = self.dataframe.iloc[index]
        impath = sample['Filepath']

        try:
            image = imread(self.root + impath)

            if len(image.shape) == 2:
                image = gray2rgb(image)
            image = img_as_float32(image)

            ilabel = labeldict[sample['Emotion']]
            label = np.array([ilabel], dtype=int)
            sample = {
                "image": image,
                "label": label,
            }

            if self.transform:
                sample = self.transform(sample)

            return sample
        except FileNotFoundError:
            print(f"File not found: {self.root + impath}. Skipping to the next record.")
            return self.__getitem__(index + 1) 
```

```python
class FaceTransform(object):
    def __init__(self, output_size, training=False):
        self.output_size = output_size
        self.training = training

    def __call__(self, sample):
        image, label = sample['image'], sample['label']
        im2 = image.copy()

        if self.training:
            if np.random.rand() > 0.5:
                im2 = np.fliplr(im2)
            if np.random.rand() > 0.5:
                im2 = np.flipud(im2)
            if np.random.rand() > 0.5:
                im2 = rotate(im2, angle=np.random.uniform(-10, 10))
            if np.random.rand() > 0.5:
                im2 = im2 * np.random.normal(1., 0.05)
            if np.random.rand() > 0.5:
                im2 = im2 + np.random.uniform(-.1, .1)
            if np.random.rand() > 0.7:
                im2 = random_noise(im2, mode='poisson')

        im2 = resize_local_mean(im2, self.output_size)
        im2 = np.transpose(im2, (2, 0, 1))

        return {
            "image": torch.from_numpy(im2).float(),
            "label": torch.from_numpy(label),
        }
```

## Modelo

```python
cfg = {
	'VGG11': [64, 'M', 128, 'M', 256, 256, 'M', 512, 512, 'M', 512, 512, 'M'],
	'VGG13': [64, 64, 'M', 128, 128, 'M', 256, 256, 'M', 512, 512, 'M', 512, 512, 'M'],
	'VGG16': [64, 64, 'M', 128, 128, 'M', 256, 256, 256, 'M', 512, 512, 512, 'M', 512, 512, 512, 'M'],
	'VGG19': [64, 64, 'M', 128, 128, 'M', 256, 256, 256, 256, 'M', 512, 512, 512, 512, 'M', 512, 512, 512, 512, 'M'],
}
class ShortVGG(nn.Module):
	def __init__(self, vgg_name='VGG19', numclasses=8, pretrain=False):
		super(ShortVGG, self).__init__()
		self.Name = vgg_name
		self.features = self._make_layers(cfg[vgg_name])
		self.avgpool = nn.AvgPool2d(kernel_size=1, stride=1)
		self.classifier = nn.Linear(512, numclasses)
		if pretrain:
			self.load_pretrain()

	def forward(self, x):
		out = self.features(x)
		out = self.avgpool(out)
		out = out.view(out.size(0), -1)
		out = F.dropout(out, p=0.5, training=self.training)
		out = self.classifier(out)
		return out

	def _make_layers(self, cfg):
		layers = []
		in_channels = 3
		for x in cfg:
			if x == 'M':
				layers += [nn.MaxPool2d(kernel_size=2, stride=2)]
			else:
				layers += [nn.Conv2d(in_channels, x, kernel_size=3, padding=1),
									nn.BatchNorm2d(x),
									nn.ReLU(inplace=True)]
				in_channels = x
		
		return nn.Sequential(*layers)
```

```python
model = ShortVGG(numclasses=len(labelnames))
criterion = nn.CrossEntropyLoss()
optimizer = Adam(model.parameters(), lr=1e-4)
```

```python
train_size = len(train_iter.dataset)
valid_size = len(valid_iter.dataset)
model.to("cuda")
for epoch in range(n_epochs):
    model.train()

    train_pbar = tqdm(total=len(train_iter), desc=f'Epoch {epoch+1:2d}/{n_epochs:2d}')
    train_acc, valid_acc = 0, 0
    loss_print = 0.
    for sample in train_iter:
        train_pbar.update(n=1)
        data = sample['image'].cuda()
        label = sample['label'].flatten().cuda()
        label = label.to(torch.int64)
        #print("Tipo de dato de label:", label.dtype)
        optimizer.zero_grad()
        
        pred = model(data)
        #print("Tipo de dato de pred:", pred.dtype)
        
        loss = criterion(pred, label)

        loss.backward()
        optimizer.step()

        train_acc += (pred.argmax(1) == label).type(torch.float).sum().item()
        loss_print += loss.item()
        train_pbar.set_postfix_str(f"Loss = {loss_print/len(train_iter):.4f}\t Train Accuracy = {train_acc/train_size:.4f}", refresh=True)
    
    train_pbar.close()
    valid_pbar = tqdm(total=len(valid_iter), desc=f'Validacion Epoch {epoch+1:2d}')

    all_labels, all_preds = np.empty((0)), np.empty((0, len(labelnames)))
    loss_print = 0.
    with torch.no_grad():
        for sample in valid_iter:
            valid_pbar.update(n=1)
            
            data = sample['image'].to('cuda')
            label = sample['label'].flatten().to('cuda')
            label = label.to(torch.int64)
            #print("Tipo de dato de label:", label.dtype)
            pred = model(data)
            #print("Tipo de dato de pred:", pred.dtype)
            loss = criterion(pred, label)

            valid_acc += (pred.argmax(1) == label).type(torch.float).sum().item()
            loss_print += loss.item()
            all_labels = np.append(all_labels, sample['label'].flatten(), axis=0)
            all_preds = np.append(all_preds, pred.cpu().numpy(), axis=0)

            valid_pbar.set_postfix_str(f"Val Loss = {loss_print/len(valid_iter):.4f} Validation Accuracy = {valid_acc/valid_size:.4f}", refresh=True)
    
    valid_pbar.close()
    f1_per_class = metrics.f1_score(all_labels, all_preds.argmax(-1), average='weighted')
    print(f"F1 = {f1_per_class}")
```

```python
import pandas as pd
import os
import torch
from torch.utils.data import Dataset

class AudioDataset(Dataset):
    def __init__(self, csvfile: str, transform=None, kind:str="Train", rootpath:str="/", root=""):
        super(AudioDataset, self).__init__()
        
        # Lee el archivo CSV en un DataFrame de Pandas.
        df0 = pd.read_csv(csvfile, sep=',', index_col='Index')
        df = df0[df0['Audio_Filter'] == 1].copy() 
        df['filepath'] = ruta_BD + "\\" + df['KEY']
        # Porcentaje deseado (10%)
#         percentage = 0.10
#         # Utiliza el método 'sample' para obtener una muestra aleatoria del 10%
#         df = df.sample(frac=percentage, random_state=42)
        existing_files = [os.path.join(root, impath) for impath in df['filepath'].tolist() if os.path.exists(os.path.join(root, impath))]

        df = df[df['filepath'].isin(existing_files)]
        
        self.dataframe = df
        self.transform = transform
        self.type = kind
        self.rootpath = rootpath
        self.sampsize = 22400 * 7 # 7 segundos 

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        sample = self.dataframe.iloc[idx]

        filepath = sample['filepath']
        duration = sample['Duration']
        label = labeldict[sample['Emotion']]

        sample = {
            "filepath": filepath,
            "duration": duration,
            "label": label,
        }

        if self.transform:
            sample = self.transform(sample)

        return sample
```

```python
class AudioTransform(object):
    def __init__(self, mode='valid'):
        self.mode = mode  
        self.sampsize = 22400 * 7  
        self.cmap = colormaps["jet"]  
        
    def __call__(self, sample):
        filepath, label, duration = sample['filepath'], sample['label'], sample['duration']

        if duration >= 7.:
            t_crop = np.random.uniform(0, duration - 7)
            padded_wav, _ = librosa.load(path=filepath, sr=22400, offset=t_crop, duration=7.)
        else:
            sampled_wav, _ = librosa.load(path=filepath, sr=22400)
            padded_wav = librosa.util.pad_center(data=sampled_wav, size=self.sampsize, mode='constant')

        if self.mode == 'train':
            if np.random.uniform() > 0.5:
                padded_wav += np.random.normal(0, 1e-5, size=padded_wav.shape)
            if np.random.uniform() > 0.5:
                rand_roll = np.random.randint(-2240, 2241)
                padded_wav = np.roll(padded_wav, rand_roll)

        spect = librosa.feature.melspectrogram(
            y=padded_wav,
            sr=22400,
            n_fft=1024,
            hop_length=self.sampsize // 255,
            n_mels=256,
        )
        spect_db = librosa.power_to_db(spect, ref=np.max, top_db=80)
        norm_spect = self.cmap(Normalize()(spect_db))

        spectrogram = np.transpose(norm_spect[..., :-1], (2, 0, 1))

        return {
            'spectrogram': torch.from_numpy(spectrogram).float(),
            'label': torch.from_numpy(np.array(label))
        }
```

```python
train_size = len(train_iter.dataset)
valid_size = len(valid_iter.dataset)

for epoch in range(n_epochs):
    tqdm_train = tqdm(
        total=len(train_iter), 
        desc=f'Train [{epoch+1}/{n_epochs}]',
    )

    train_acc, valid_acc = 0, 0

    model.train()
    for sample in train_iter:
        tqdm_train.update(n=1)

        torch.cuda.empty_cache()
        
        data = sample['spectrogram'].to("cuda")
        label = sample['label'].to(torch.int64).to("cuda")

        pred = model(data)
        loss = loss_fn(pred, label)

        opt_fn.zero_grad()
        loss.backward()
        opt_fn.step()

        train_acc += (pred.argmax(1) == label).type(torch.float).sum().item() / train_size
        tqdm_train.set_postfix_str(
            f'Loss={loss.item():.4f} - Train Accuracy={train_acc:.4f}',
            refresh=True,
        )
    
    tqdm_train.close()

    tqdm_val = tqdm(total=len(valid_iter), desc=f'Validation')

    all_labels, all_preds = np.empty((0)), np.empty((0, len(labelnames)))

    model.eval()
    with torch.no_grad():
        for sample in valid_iter:
            tqdm_val.update(n=1)

            torch.cuda.empty_cache()
            
            data = sample['spectrogram'].to("cuda")
            label = sample['label'].to(torch.int64).to("cuda")

            pred = model(data)
            loss = loss_fn(pred, label)

            valid_acc += (pred.argmax(1) == label).type(torch.float).sum().item() / valid_size
            all_labels = np.append(all_labels, label.cpu().data, axis=0)
            all_preds = np.append(all_preds, pred.cpu().data, axis=0)

            tqdm_val.set_postfix_str(
                f'Val Loss={loss.item():.4f} - Validation Accuracy={valid_acc:.4f}',
                refresh=True
            )
            
    librosa.cache.clear()
    tqdm_val.close()
    f1_per_class = metrics.f1_score(all_labels, all_preds.argmax(-1), average=None)
    f1_all = metrics.f1_score(all_labels, all_preds.argmax(-1), average='macro')
    
    display_pretty(f"Epoch {epoch+1} = f1 score={f1_all:.4f}")
```

## Carga de datos

```python
def open_meld(csvname):
    df = pd.read_csv(csvname, sep=',', index_col='Index')
    df = df[df['Text_Filter'] == 1].copy()
    # Porcentaje deseado (10%)
#     percentage = 0.10
#     # Utiliza el método 'sample' para obtener una muestra aleatoria del 10%
#     df = df.sample(frac=percentage, random_state=42)
    return df

trainset = open_meld(train_file)
validset = open_meld(valid_file)

trainset.head()
```

```python
Ytrain = trainset.Emotion.map(lambda x: labeldict[x], na_action='ignore')
Yvalid = validset.Emotion.map(lambda x: labeldict[x], na_action='ignore')
```

## Entrenamiento

```python
# Embedding de texto
bertmodels = 'bert-large-nli-stsb-mean-tokens'
model = SentenceTransformer(bertmodels)
```

```python
train_text = trainset.Lemmatized_Tokens.to_list()
valid_text = validset.Lemmatized_Tokens.to_list()
```

```python
train_text = [str(text) for text in trainset.Lemmatized_Tokens if isinstance(text, str)]
valid_text = [str(text) for text in validset.Lemmatized_Tokens if isinstance(text, str)]
```

```python
train_embed = model.encode(train_text)
valid_embed = model.encode(valid_text)
```

```python
print(train_text)
```

```python
clf = MLPClassifier(
    hidden_layer_sizes=(500, 300,), 
    learning_rate_init=learning_rate,
    tol=1e-6, 
    early_stopping=True,
    verbose=True,
)

clf.fit(train_embed, Ytrain.tolist())
```

```python
pred = clf.predict_proba(valid_embed)
```

```python
cm = metrics.confusion_matrix(Yvalid, pred.argmax(-1))
print((cm.diagonal()/cm.sum(1))*100)
print((cm.diagonal().sum()/cm.sum())*100)
```

```python
print(metrics.classification_report(Yvalid, pred.argmax(-1), target_names=labelnames, digits=3))
```
