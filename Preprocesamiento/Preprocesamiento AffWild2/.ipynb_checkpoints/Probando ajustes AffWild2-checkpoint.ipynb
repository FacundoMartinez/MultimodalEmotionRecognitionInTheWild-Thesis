```python
import os
import pandas as pd
from natsort import natsorted

carpeta = r"D:\sentirobots\Datasets\AffWild2\Audios_AffWild2\Audios cortados"

nombres_archivos = []
rutas_completas = []

for nombre_archivo in os.listdir(carpeta):
    ruta_completa = os.path.join(carpeta, nombre_archivo)
    if os.path.isfile(ruta_completa):
        nombres_archivos.append(nombre_archivo)
        rutas_completas.append(ruta_completa)

nombres_archivos = natsorted(nombres_archivos)
rutas_completas = natsorted(rutas_completas)
data = {'NombreArchivo': nombres_archivos, 'RutaCompleta': rutas_completas}
df = pd.DataFrame(data)

df
```

```python
df['carpeta'] = df['NombreArchivo'].str.split('.').str[0]

df
```

```python
registro_por_posicions = df.iloc[2][0]
print(registro_por_posicions)
registro_por_posicion = df.iloc[2][1]
print(registro_por_posicion)
```

```python
df_original = pd.read_csv(r'D:\sentirobots\Datasets\AffWild2\Audios_AffWild2\emociones_AffWild2_train_val_Audios.csv')
df_original
```

```python
df_original['nombre_sin_ext'] = df_original['nombre_archivo'].str.split('.').str[0]
df_original
```

```python
current_name = ""
counter = 0
id_list = []

for index, row in df_original.iterrows():
    if row['nombre_sin_ext'] == current_name:
        counter += 1
    else:
        counter = 1
        current_name = row['nombre_sin_ext']
    id_list.append(counter)

df_original['id'] = id_list
df_original
```

```python
df_original['carpeta'] = df_original['nombre_sin_ext'] + '_' + df_original['id'].astype(str)

df_original
```

```python
result = pd.merge(df, df_original, on='carpeta', how='inner')
result
```

```python
import os
import pandas as pd

# Ruta base
ruta_base = r'D:\sentirobots\Datasets\AffWild2\cropped_images\caras'

data = {
    'carpeta_1': ['1-30-1280x720', '2-30-640x360', '3-25-1920x1080'],
    'carpeta_2': ['carpeta4', 'carpeta5', 'carpeta6'],
    'carpeta_3': ['carpeta7', 'carpeta8', 'carpeta9']
}
df = pd.DataFrame(data)

def obtener_nombres_archivo(carpeta):
    carpeta_completa = os.path.join(ruta_base, carpeta)
    nombres_archivo = []
    for archivo in os.listdir(carpeta_completa):
        archivo_completo = os.path.join(carpeta_completa, archivo)
        if os.path.isfile(archivo_completo):
            nombre_sin_extension = os.path.splitext(archivo)[0]
            nombres_archivo.append(nombre_sin_extension)
    return nombres_archivo

df['nombres_archivo'] = df.apply(lambda row: obtener_nombres_archivo(row['carpeta_1']), axis=1)
df['carpeta_completa'] = df.apply(lambda row: os.path.join(ruta_base, row['carpeta_1']), axis=1)
df
```
