```python
import pandas as pd

archivos = [r'D:\sentirobots\Datasets\AffWild2\AffWild2-Face-Parte1.csv', 
            r'D:\sentirobots\Datasets\AffWild2\AffWild2-Face-Parte11.csv', 
            r'D:\sentirobots\Datasets\AffWild2\AffWild2-Face-Parte2.csv', 
            r'D:\sentirobots\Datasets\AffWild2\AffWild2-Face-Parte22.csv',
            r'D:\sentirobots\Datasets\AffWild2\AffWild2-Face-Parte3.csv',
            r'D:\sentirobots\Datasets\AffWild2\AffWild2-Face-Parte4.csv']

dataframes = []

for archivo in archivos:
    df = pd.read_csv(archivo)
    dataframes.append(df)
dataframe_concatenado = pd.concat(dataframes, ignore_index=True)
dataframe_concatenado
```

```python
valor = dataframe_concatenado.loc[0, 'RutaArchivo']  # Obtiene el valor en la fila 1 y columna 'nombres_archivo'
print(valor)
```

```python
import os

def obtener_ultima_carpeta(ruta):
    carpetas = ruta.split('\\')
    return carpetas[-2] if len(carpetas) >= 2 else None

dataframe_concatenado['ultima_carpeta'] = dataframe_concatenado['RutaArchivo'].apply(obtener_ultima_carpeta)

dataframe_concatenado
```

```python
import locale

def obtener_nombre_archivo(ruta):
    return os.path.basename(ruta)

dataframe_concatenado['nombre_archivo'] = dataframe_concatenado['RutaArchivo'].apply(obtener_nombre_archivo)

locale.setlocale(locale.LC_COLLATE, 'en_US.UTF-8')

# Crear una nueva columna 'orden_temp' utilizando locale.strxfrm
dataframe_concatenado['orden_temp'] = dataframe_concatenado['ultima_carpeta'].apply(locale.strxfrm)

# Ordenar el DataFrame segÃºn la columna 'orden_temp' pero mantener el Ã­ndice original
df_sorted = dataframe_concatenado.sort_values(by='orden_temp').reset_index(drop=True)

# Eliminar la columna 'orden_temp'
df_sorted.drop('orden_temp', axis=1, inplace=True)

df_sorted
```

```python
valor = df_sorted.loc[0, 'RutaArchivo']  # Obtiene el valor en la fila 1 y columna 'nombres_archivo'
print(valor)
```

```python
def ordenar_grupo(group):
    return group.sort_values(by='nombre_archivo', ascending=True)

nuevo_df = df_sorted.groupby('ultima_carpeta', group_keys=False).apply(ordenar_grupo)
nuevo_df
```

```python
#nuevo_df.drop('orden_temp_ultima_carpeta', axis=1, inplace=True)
nuevo_df
```

```python
nuevo_df["KEY"] = nuevo_df["ultima_carpeta"] + "\\" + nuevo_df["nombre_archivo"]
nuevo_df
```

```python
nuevo_df.to_csv(r'D:\sentirobots\Datasets\AffWild2\Face_AffWild2.csv', index=False)
```
