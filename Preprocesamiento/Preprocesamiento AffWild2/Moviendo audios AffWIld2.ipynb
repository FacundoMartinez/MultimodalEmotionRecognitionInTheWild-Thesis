```python
import os
import pandas as pd
import shutil

ruta_origen = r"D:\sentirobots\Datasets\AffWild2\Audios_AffWild2\Seleccion"
ruta_destino_train = r"D:\sentirobots\Datasets\Base de Datos Unificada\Training\Audios"
ruta_destino_val = r"D:\sentirobots\Datasets\Base de Datos Unificada\Validation\Audios"

ruta_csv = r"D:\sentirobots\Datasets\etiquetas-unificadas.csv"

data = pd.read_csv(ruta_csv)

for index, row in data.iterrows():
    nombre_archivo = row["Nombre_Archivo"]
    tipo = row["Tipo"]
    origen_completo = os.path.join(ruta_origen, f"{nombre_archivo}.wav")
    
    if tipo == "train":
        destino_completo = os.path.join(ruta_destino_train, f"{nombre_archivo}.wav")
    elif tipo == "val":
        destino_completo = os.path.join(ruta_destino_val, f"{nombre_archivo}.wav")
    else:
        print(f"Tipo no reconocido para el archivo {nombre_archivo}")
        continue

    try:
        shutil.move(origen_completo, destino_completo)
        print(f"Archivo {nombre_archivo}.wav movido exitosamente a {destino_completo}")
    except FileNotFoundError:
        print(f"Archivo {nombre_archivo}.wav no encontrado en la ruta de origen.")
    except Exception as e:
        print(f"Error al mover el archivo {nombre_archivo}.wav: {e}")
```
