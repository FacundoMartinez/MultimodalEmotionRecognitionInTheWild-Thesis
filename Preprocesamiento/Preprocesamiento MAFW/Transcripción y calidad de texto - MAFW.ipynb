```python
import pandas as pd
import speech_recognition as sr
from http.client import IncompleteRead
```

```python
def transcripcion(audio_path):
    r = sr.Recognizer()
    with sr.AudioFile(audio_path) as source:
        audio = r.record(source)

        try:
            # Utilizar Google Speech Recognition para realizar la transcripción
            text = r.recognize_google(audio)
            return text
        except sr.UnknownValueError:
            print("No se pudo reconocer el audio")
            return ""
        except sr.RequestError as e:
            print("Error al hacer la solicitud al servicio de reconocimiento de voz: ", e)
            return ""
        except IncompleteRead:
            # Manejar el caso en el que se produzca una lectura incompleta de la respuesta HTTP
            print("Error: Lectura incompleta de la respuesta HTTP.")
            return ""
```

```python
df = pd.read_csv(r'D:\sentirobots\Datasets\MAFW\MAFW-20230614T154044Z-003\MAFW\data\Filtro de audios - MAFW\filtro-audio-MAFW.csv')

df['Transcripcion'] = ''
df
```

```python
cont=0
for index, row in df.iterrows():
    if row['Filtro2'] == 1:
        cont=cont+1
        print(cont)
        ruta_archivo = row['RutaArchivo']
        transcripcion_audio = transcripcion(ruta_archivo)
        df.at[index, 'Transcripcion'] = transcripcion_audio
```
