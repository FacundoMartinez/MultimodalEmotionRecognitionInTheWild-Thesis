```python
import pandas as pd
import speech_recognition as sr
```

```python
def transcripcion(audio_path):
    r = sr.Recognizer()
    with sr.AudioFile(audio_path) as source:
        audio = r.record(source)
        try:
            # Utilizar Google Speech Recognition para realizar la transcripción
            text = r.recognize_google(audio)
            return text
        except sr.UnknownValueError:
            print("No se pudo reconocer el audio")
            return ""
        except sr.RequestError as e:
            print("Error al hacer la solicitud al servicio de reconocimiento de voz: ", e)
            return ""
```

```python
df = pd.read_csv(r'D:\sentirobots\Datasets\AFEW\Datos\Filtro de audios - AFEW\filtro-audio-AFEW.csv')

df['Transcripcion'] = ''
df
```

```python
cont=0
for index, row in df.iterrows():
    if row['Filtro2'] == 1:
        ruta_archivo = row['RutaArchivo']
        transcripcion_audio = transcripcion(ruta_archivo)
        df.at[index, 'Transcripcion'] = transcripcion_audio
```

```python
import re
import emoji

def quitar_caracteres_especiales(texto):
    # Definir el patrón de caracteres especiales utilizando una expresión regular
    patron = r'[^\w\s?!.,:;\'"]'
    texto_sin_especiales = re.sub(patron, '', texto)
    texto_sin_emojis = emoji.demojize(texto_sin_especiales)
    patron = re.compile(r':[^:\s]*(?:::[^:\s]*)*:')
    texto_limpio = re.sub(patron, '', texto_sin_emojis)    
    return texto_limpio
#texto = "¡Hola, @usuario! ¿Cómo estás? #Python es genial!!! Me dijo: Cómo te llamas; luego de haber comido 😃🚀"
#clean_text = quitar_caracteres_especiales(texto)
#print(clean_text)
```

```python
from speechbrain.pretrained import EncoderDecoderASR
```

```python
import librosa
import torch
from tqdm import tqdm
os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"
model = EncoderDecoderASR.from_hparams(
    source="speechbrain/asr-transformer-transformerlm-librispeech",
    savedir="pretrained_models/asr-transformer-transformerlm-librispeech",
)
sample_rate = 22400
max_time = 7.

texts = []
# Crear una nueva columna en el DataFrame para almacenar las transcripciones
df['SpeechBrain'] = ''

for index, row in tqdm(df.iterrows(), total=len(df)):
    filepath = row['RutaArchivo']
    # Cargamos el audio y hacemos padding
    sampled_wav, _ = librosa.load(filepath, sr=sample_rate)
    total_length = int(sample_rate * max_time)
    padded_wav = librosa.util.pad_center(data=sampled_wav, size=total_length, mode='constant')
    normalized = model.audio_normalizer(torch.tensor(padded_wav), sample_rate)
    rel_length = torch.tensor([1.])
    predicted_words, _ = model.transcribe_batch(
        normalized.unsqueeze(0),
        rel_length,
    )
    normalized = normalized.detach()
    rel_length = rel_length.detach()
    torch.cuda.empty_cache()

    df.at[index, 'SpeechBrain'] = " ".join(predicted_words)
```
