# Detección de rango de voz

```python
import os
import pandas as pd
import librosa
from speechbrain.pretrained import EncoderDecoderASR
from tqdm import tqdm
import numpy as np
import torch
```

### Crear df desde el csv

```python
rutaa= r"D:\sentirobots\filtro-audio-pruebas.csv"
df = pd.read_csv(rutaa)
df
```

### Obtener duraciones de los audios

```python
def obtener_duracion_audio(ruta_archivo):
    audio, sr = librosa.load(ruta_archivo)
    duracion = len(audio) / sr
    return duracion

df['Duracion'] = df['RutaArchivo'].apply(obtener_duracion_audio)
df["Duracion"]
```

### Listado de frecuencias

```python
import numpy as np
import librosa
import pandas as pd

def analyze_audio_frequencies(audio_path):
    data, sr = librosa.load(audio_path, sr=None)
    Audio_m = data
    L = len(Audio_m)
    n = np.arange(0, L) / sr
    gk = np.fft.fft(Audio_m)
    M_gk = abs(gk)
    M_gk = M_gk[0:L // 2]
    Ph_gk = np.angle(gk)
    F = sr * np.arange(0, L // 2) / L
    return F

df["Frecuencias"] = df["RutaArchivo"].apply(analyze_audio_frequencies)
df["Frecuencias"]
```

### Convertir las frecuencias a numeros enteros

```python
def convertir_lista_a_enteros(lista):
    return [int(valor) for valor in lista]

df['Frecuencias'] = df['Frecuencias'].apply(convertir_lista_a_enteros)
df['Frecuencias']
```

### Frecuencia dominante

```python
def analyze_audio_frequency(audio_path):
    data, sr = librosa.load(audio_path, sr=None)
    Audio_m = data
    L = len(Audio_m)
    n = np.arange(0, L) / sr
    gk = np.fft.fft(Audio_m)
    M_gk = abs(gk)
    M_gk = M_gk[0:L // 2]
    Ph_gk = np.angle(gk)
    F = sr * np.arange(0, L // 2) / L
    voice_range = (F >= 0) & (F <= 10000000)
    filtered_frequencies = F[voice_range]
    filtered_magnitudes = M_gk[voice_range]
    dominant_frequency = filtered_frequencies[np.argmax(filtered_magnitudes)]
    return dominant_frequency

df["Frecuencia dominante"] = df["RutaArchivo"].apply(analyze_audio_frequency)
df["Frecuencia dominante"]
```

### Media y desv. estandar

```python
import numpy as np

def calculate_mean_std(data):
    positive_values = [x for x in data if x > 0]
    mean = np.mean(positive_values)
    std = np.std(positive_values)
    return pd.Series([mean, std])

df[["Mean", "Std"]] = df["Frecuencias"].apply(calculate_mean_std)
df[["Mean", "Std"]]
```

### Contar cuantas frecuencias se encuentran en el rando de voz humana (entre 100Hz y 4.000Hz)

```python
rango_min = 100
rango_max = 4000

def contar_en_rango(lista):
    return sum(1 for num in lista if rango_min <= num <= rango_max)

df['CantidadEnRango'] = df['Frecuencias'].apply(contar_en_rango)
```

### Contar cuantas frecuencias hay en total

```python
def contar_sin_rango(lista):
    return sum(1 for num in lista)

df['CantidadSinRango'] = df['Frecuencias'].apply(contar_sin_rango)
```

### % de frecuencia

```python
def divide_columns(row):
    return row["CantidadEnRango"] / row["CantidadSinRango"]*100

df["% frec. voz"] = df.apply(divide_columns, axis=1)
df
```

# SpeechBrain

```python
model = EncoderDecoderASR.from_hparams(
    source="speechbrain/asr-transformer-transformerlm-librispeech",
    savedir="pretrained_models/asr-transformer-transformerlm-librispeech",
#    run_opts={'device':'cuda'},
)
sample_rate = 22400
max_time = 7.

texts = []
```

```python
df['Text'] = ''

for index, row in tqdm(df[df['Filtro2'] < 3].iterrows()):
    audiofile, _ = librosa.load(path=row['RutaArchivo'], sr=22400, duration=np.minimum(row['Duracion'], 7.))

    if row['Duracion'] < 7.:
        audiofile = librosa.util.pad_center(data=audiofile, size=22400*7)

    normalized = model.audio_normalizer(torch.tensor(audiofile), 22400)
    rel_length = torch.tensor([1.])
    predicted_words, _ = model.transcribe_batch(
        normalized.unsqueeze(0),
        rel_length,
    )

    normalized = normalized.detach()
    rel_length = rel_length.detach()
    torch.cuda.empty_cache()

    df.at[index, 'Text'] = " ".join(predicted_words)

df
```

```python
columna = "Text"
num_vacios = df[columna].eq('').sum()
total_registros = df[columna].shape[0]
porcentaje_vacios = (num_vacios / total_registros) * 100

print("Número de valores vacíos en la columna {}: {}".format(columna, num_vacios))
print("Total de registros en la columna {}: {}".format(columna, total_registros))
print("Porcentaje de valores vacíos en la columna {}: {:.2f}%".format(columna, porcentaje_vacios))
```
