# Calculando HOG CON SIMILITUD DEL COSENO

```python
import os
import pandas as pd

ruta_carpeta = r'D:\Datasets\MERDWild\Training\Rostros'

rutas_completas = []

for root, dirs, files in os.walk(ruta_carpeta):
    for file in files:
        ruta_completa = os.path.join(root, file)
        rutas_completas.append(ruta_completa)
        
img = pd.DataFrame({'Ruta Completa': rutas_completas})

img
```

```python
def obtener_key(ruta):
    partes = ruta.split(os.sep)[-4:]  
    key = os.sep.join(partes)  
    return key

img['KEY'] = img['Ruta Completa'].apply(obtener_key)

img
```

```python
def obtener_folder(ruta):
    partes = ruta.split(os.sep)
    folder = partes[-2] 
    return folder

img['Folder'] = img['Ruta Completa'].apply(obtener_folder)
img
```

```python
cambios = img['Folder'] != img['Folder'].shift(1)
indices_cambios = cambios.index[cambios].tolist()

total_cambios = sum(cambios)

indice_10 = indices_cambios[int(0.10 * total_cambios)]
indice_20 = indices_cambios[int(0.20 * total_cambios)]
indice_30 = indices_cambios[int(0.30 * total_cambios)]
indice_40 = indices_cambios[int(0.40 * total_cambios)]
indice_50 = indices_cambios[int(0.50 * total_cambios)]
indice_60 = indices_cambios[int(0.60 * total_cambios)]
indice_70 = indices_cambios[int(0.70 * total_cambios)]
indice_80 = indices_cambios[int(0.80 * total_cambios)]
indice_90 = indices_cambios[int(0.90 * total_cambios)]

print("Índice correspondiente al 20% de cambios:", indice_10)
print("Índice correspondiente al 40% de cambios:", indice_20)
print("Índice correspondiente al 60% de cambios:", indice_30)
print("Índice correspondiente al 80% de cambios:", indice_40)
print("Índice correspondiente al 20% de cambios:", indice_50)
print("Índice correspondiente al 40% de cambios:", indice_60)
print("Índice correspondiente al 60% de cambios:", indice_70)
print("Índice correspondiente al 80% de cambios:", indice_80)
print("Índice correspondiente al 20% de cambios:", indice_90)
```

```python
RESPALDO_IMAGEN=img.copy()
inicio = 0
fin = 96444
RESPALDO_IMAGEN = RESPALDO_IMAGEN.iloc[inicio:fin].copy()
RESPALDO_IMAGEN
```

```python
import skimage
from skimage.feature import hog
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def sim_coseno(ruta_imagen1, ruta_imagen2):
    # Cargar las imágenes en escala de grises
    image1 = skimage.io.imread(ruta_imagen1, as_gray=True)
    image2 = skimage.io.imread(ruta_imagen2, as_gray=True)

    # Obtener los descriptores de características HOG
    fd1 = hog(image1, orientations=8, pixels_per_cell=(16, 16),
              cells_per_block=(1, 1), block_norm='L2-Hys', feature_vector=True)
    fd2 = hog(image2, orientations=8, pixels_per_cell=(16, 16),
              cells_per_block=(1, 1), block_norm='L2-Hys', feature_vector=True)

    # Calcular la similitud del coseno entre los descriptores
    similitud = cosine_similarity([fd1], [fd2])

    return similitud[0][0]
```

```python
similitudes = []

for grupo, df_grupo in RESPALDO_IMAGEN.groupby('Folder'):
    rutas_grupo = df_grupo['Ruta Completa'].tolist()
    n = len(rutas_grupo)
    
    for i in range(n):
        ruta_imagen1 = os.path.normpath(rutas_grupo[i])
        
        if i == n - 1:
            # Si estamos en el último registro de la carpeta, usar la similitud calculada previamente
            sim = similitudes[-1]
        else:
            ruta_imagen2 = os.path.normpath(rutas_grupo[i + 1])
            sim = sim_coseno(ruta_imagen1, ruta_imagen2)
            
        similitudes.append(sim)

RESPALDO_IMAGEN['Similitud'] = similitudes
```
