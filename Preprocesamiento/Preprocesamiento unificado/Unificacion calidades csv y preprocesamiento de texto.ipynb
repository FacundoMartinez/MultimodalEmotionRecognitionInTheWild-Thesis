```python
import os
def obtener_tipo(ruta):
    partes = ruta.split(os.sep)
    if len(partes) >= 2:
        return partes[-4]
    else:
        return None

archivo3["Tipo"] = archivo3["RutaArchivo"].apply(obtener_tipo)
archivo3['Tipo'] = archivo3['Tipo'].replace({'Val': 'val', 'Train': 'train'})
archivo3
```

```python
def cambiar_valores(row):
    if pd.notnull(row["resolucion"]):
        if row["resolucion"] == True:
            row["resolucion H"] = 70
            row["resolucion W"] = 70
        else:
            row["resolucion H"] = 30
            row["resolucion W"] = 30
    return row

df_concatenado = df_concatenado.apply(cambiar_valores, axis=1)
df_concatenado
```

```python
import pandas as pd
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()  # Dividir el texto en palabras
    filtered_words = [word for word in words if word.lower() not in stop_words]  # Eliminar stopwords
    return ' '.join(filtered_words)  # Unir las palabras nuevamente

DF_FINAL_AUDIO['Texto_without_stopwords'] = DF_FINAL_AUDIO['Transcripcion2'].apply(remove_stopwords)
DF_FINAL_AUDIO
```

```python
import pandas as pd
import spacy

with open(r'C:\Users\Desarrollo\Desktop\English_Words.txt', 'r') as file:
    english_words = set(word.strip().lower() for word in file)

nlp = spacy.load("en_core_web_sm")

def reduce_to_existing_word(word):
    original_word = word 
    
    if word.lower() in english_words:
        return word.lower() 
    
    reduced_word = word
    iterations = 0 
    max_iterations = 10
    while len(reduced_word) > 1 and iterations < max_iterations:
        reduced_word = re.sub(r'(\w)(\1{2,})', r'\1\1', reduced_word)
        print(reduced_word)
        if reduced_word.lower() in english_words:
            return reduced_word.lower()
        
        if reduced_word == word:
            break
        iterations += 1
        reduced_word = word
        iterations = 0 
        max_iterations = 10  
        while len(reduced_word) > 1 and iterations < max_iterations:
            reduced_word = re.sub(r'(\w)\1{1,}', r'\1\1', reduced_word)
            if reduced_word.lower() in english_words:
                return reduced_word.lower() 
            if reduced_word == word:
                break
            iterations += 1
        break
    return original_word

def lemmatize_tokens(tokens):
    text = ' '.join(tokens)
    doc = nlp(text)
    lemmatized_tokens = [token.lemma_ for token in doc]
    return lemmatized_tokens

DF_FINAL_AUDIO['Fila'] = range(1, len(DF_FINAL_AUDIO) + 1)

results = []

for index, row in DF_FINAL_AUDIO.iterrows():
    print(f"Procesando fila {row['Fila']}...")
    tokens = row['Tokens']
    lemmatized_tokens = ['[]'] if not tokens else lemmatize_tokens([reduce_to_existing_word(word) for word in tokens])
    results.append(lemmatized_tokens)

DF_FINAL_AUDIO['Lemmatized_Tokens'] = results
DF_FINAL_AUDIO[['Fila', 'Texto_without_stopwords', 'Tokens', 'Lemmatized_Tokens']]
```
